```python
import pandas as pd
```

```python
dtype = {
    'Year': str
}
raw17 = pd.read_csv('raw/MI_LivestockCountByCounty_2017.csv', dtype=dtype)
raw18 = pd.read_csv('raw/MI_LivestockCountByCounty_2018.csv', dtype=dtype)

# drop total line at end
raw17 = raw17[raw17['County Id'] != 'TOTALS']
raw18 = raw18[raw18['County Id'] != 'TOTALS']
```

## Example script showing how livestock data was tidied

```python
# Note raw data comes in "wide"/non-tidy format
raw17.head()
```

```python
# Using Pandas "melt" function to shift data to "tidy" format

tall17 = raw17.melt(
    id_vars=['Year','County Id','County Name'],
    var_name='Type',
    value_name='Count',
).fillna(0)

tall18 = raw18.melt(
    id_vars=['Year','County Id','County Name'],
    var_name='Type',
    value_name='Count',
).fillna(0) # fillna(0) replaces missing values with 0
```

```python
tall17.head()
```

```python
# Combine multiple years
df = tall17.append(tall18)
```

```python
# Export
df.to_csv('data/mt-livestock-counts.csv', index=False)
# index=False keeps row numbers from cluttering up .csv

# Check CSV file in text editor/Excel/Google Sheets to make sure it's what's expected
```
